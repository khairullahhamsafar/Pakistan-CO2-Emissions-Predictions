```python
import pandas as pd
```

```python
dataset = pd.read_csv('/kaggle/input/climate-change-indicators-for-pakistan/Dataset_Clean_Climate_change_indicators_for_pakistan.csv')
```

```python
dataset.head()
```

```python
dataset  = dataset.drop(['Sno','Country Name'], axis = 1)
```

```python
dataset.isna().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(dataset)
dataset = pd.DataFrame(imputer.transform(dataset),columns = dataset.columns)
```

```python
dataset
```

```python
print(dataset.columns)
```

```python
dataset.corr().tail()
```

```python
import matplotlib.pyplot as plt
for col in dataset.columns:
    if col != 'Year':
        plt.figure(figsize=(12, 12), dpi=80)
        plt.plot(dataset['Year'],dataset[col])
        plt.title(col+" per year")
        plt.xlabel('Years')
        plt.ylabel(col)
        plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
for col in dataset.columns:
    if col != 'Year' and col!='RainFall Mean ()':
        figure(figsize=(12, 12), dpi=80)
        plt.plot(dataset['Year'],dataset[col]/dataset[col].iloc[0]*100,label = col)
        plt.plot(dataset['Year'],dataset['RainFall Mean ()']/dataset['RainFall Mean ()'].iloc[0]*100,label = 'Rainfall')
        plt.title(col+" per year")
        plt.xlabel('Years')
        plt.ylabel(col)
        plt.legend()
        plt.show()
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize = (25, 25))
sns.heatmap(dataset.corr(), annot = True)
```

```python
X = dataset.drop(['Year','RainFall Mean ()'],axis = 1)
y = dataset['RainFall Mean ()']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
def evaluate(y_test,pred):
    return {
        "R2":r2_score(y_test,pred),
        "MAE":mean_absolute_error(y_test,pred),
        "MSE":mean_squared_error(y_test,pred),
    }
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
regressors = [LinearRegression(),Ridge(),KNeighborsRegressor()]
for regressor in regressors:
    regressor.fit(X_train,y_train)
    pred = regressor.predict(X_test)
    scores = evaluate(y_test,pred)
    print("Regressor "+str(regressor))
    print(scores)
    plt.plot(list(y_test),label = 'Original')
    plt.plot(pred, label = 'Predicted')
    plt.legend
    plt.title(str(regressor)+' Original and Predicted Outp')
    plt.show()
```
